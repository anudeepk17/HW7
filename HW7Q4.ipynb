```python
import random
from prettytable import PrettyTable
```

```python
def printQ(q_matrix):
  t = PrettyTable(['Q value', 'Move', 'Stay'])
  t.add_row(['A', q_matrix[0][0],q_matrix[0][1]])
  t.add_row(['B', q_matrix[1][0],q_matrix[1][1]])
  print(t)
```

```python
environment_matrix = [[0, 1],
                      [0, 1]]
q_matrix = [[0, 0],
            [0, 0]]
def getNextState(cur_pos, action):
    if (action == 0):
        return (cur_pos+1)%2
    else:
        return cur_pos

discount = 0.8
learning_rate = 0.5
cur_pos=0
for _ in range(200):
# get starting place A=0 B=1
    #cur_pos = random.choice([0,1]
    # get all possible next states from cur_pos 0 = Move 1 = Stay
    possible_actions = [0,1]
    # select any one action
    if q_matrix[cur_pos][0]==q_matrix[cur_pos][1]:
      action = 0
    else:
      action = q_matrix[cur_pos].index(max(q_matrix[cur_pos]))
    # find the next state corresponding to the action selected
    next_state = getNextState(cur_pos, action)
    # update the q_matrix
    q_matrix[cur_pos][action] = q_matrix[cur_pos][action] + learning_rate * (environment_matrix[cur_pos][action] + 
        discount * max(q_matrix[next_state]) - q_matrix[cur_pos][action])
    cur_pos=next_state

printQ(q_matrix)
```
